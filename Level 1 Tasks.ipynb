## **Task 1: Top Cuisines**

**Determine the top three most common cuisines in the dataset.**
**Calculate the percentage of restaurants that serve each of the top cuisines.**


```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Dataset .csv'
data = pd.read_csv('Dataset .csv')

print(data.head())
```

```python
# Split the 'Cuisines' column to count individual cuisines
cuisines_series = data['Cuisines'].str.split(', ').explode()

# Count the occurrences of each cuisine
cuisine_counts = cuisines_series.value_counts()

# Get the top three cuisines
top_cuisines = cuisine_counts.head(3)

# Calculate the percentage of restaurants serving each of the top cuisines
total_restaurants = len(data)
top_cuisines_percentage = (top_cuisines / total_restaurants) * 100

# Data for visualization
cuisines_names = top_cuisines.index
cuisines_values = top_cuisines.values

# Plotting the results
plt.figure(figsize=(10, 8))
plt.bar(cuisines_names, cuisines_values, color=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Top 3 Cuisines in Restaurants')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Display the percentage on top of the bars
for i, v in enumerate(cuisines_values):
    plt.text(i, v + 5, f"{top_cuisines_percentage[i]:.2f}%", ha='center')

plt.tight_layout()
plt.show()
```



## **Task 2 :  City Analysis**

**Identify the city with the highest number of restaurants in the dataset.**
**Calculate the average rating for restaurants in each city.**
**Determine the city with the highest average rating.**


```python
# importing the libraries
import pandas as pd

# Load the dataset 
file_path = 'Dataset .csv'
data = pd.read_csv('Dataset .csv')


# Identify the city with the highest number of restaurants
city_counts = data['City'].value_counts()
city_with_most_restaurants = city_counts.idxmax()
num_restaurants = city_counts.max()


# Calculate the average rating for restaurants in each city
average_rating_per_city = data.groupby('City')['Aggregate rating'].mean()


# Determine the city with the highest average rating
city_with_highest_avg_rating = average_rating_per_city.idxmax()
highest_avg_rating = average_rating_per_city.max()

# Prepare insights as a dictionary
city_analysis_insights = {
    "City with the Most Restaurants": city_with_most_restaurants,
    "Number of Restaurants": num_restaurants,
    "City with the Highest Average Rating": city_with_highest_avg_rating,
    "Highest Average Rating": round(highest_avg_rating, 2)
}

city_analysis_insights
```

```python
# Visualization 1: Top 10 cities by number of restaurants 

# Import necessary libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
top_cities_by_restaurants = city_counts.head(10)

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=top_cities_by_restaurants.values, 
    y=top_cities_by_restaurants.index, 
    palette='pastel'
)

# Add annotations for the bar values
for index, value in enumerate(top_cities_by_restaurants.values):
    plt.text(value + 20, index, str(value), va='center', fontsize=10)  

# Title and labels
plt.title('Top 10 Cities by Number of Restaurants', fontsize=14)
plt.xlabel('Number of Restaurants', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Top 10 cities by average rating 

# Import necessary libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

top_cities_by_avg_rating = average_rating_per_city.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=top_cities_by_avg_rating.values, 
    y=top_cities_by_avg_rating.index, 
    palette='coolwarm'
)

# Add annotations for the bar values
for index, value in enumerate(top_cities_by_avg_rating.values):
    plt.text(value + 0.05, index, f"{value:.2f}", va='center', fontsize=10)  

# Title and labels
plt.title('Top 10 Cities by Average Restaurant Rating', fontsize=14)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.tight_layout()
plt.show()
```

## **Task 3 :  Price Range Distribution**
**Create a histogram or bar chart to
visualize the distribution of price ranges
among the restaurants.
Calculate the percentage of restaurants
in each price range category.**


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Dataset .csv'
dataset = pd.read_csv('Dataset .csv')

# Display the first few rows of the dataset
dataset.head()
```

```python
# Step 1: Convert the "Price range" column to numeric
dataset['Price range'] = pd.to_numeric(dataset['Price range'], errors='coerce')

# Step 2: Calculate the percentage distribution of restaurants in each price range
price_range_counts = dataset['Price range'].value_counts().sort_index()
total_restaurants = price_range_counts.sum()
price_range_percentages = (price_range_counts / total_restaurants) * 100

# Step 3: Visualize the distribution with a bar chart
plt.figure(figsize=(8, 6))
price_range_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Price Ranges Among Restaurants', fontsize=14)
plt.xlabel('Price Range', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the percentage distribution
price_range_percentages.round(2)
```

## **Task 4 : Online Delivery**
**Determine the percentage of restaurants
that offer online delivery.
Compare the average ratings of restaurants
with and without online delivery**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Dataset .csv")

# Check necessary columns
if 'Has Online delivery' not in df.columns or 'Aggregate rating' not in df.columns:
    print("Error: Required columns not found in dataset.")
else:
    # Normalize 'Has Online delivery' column
    df['Has Online delivery'] = df['Has Online delivery'].astype(str).str.lower()

    # Calculate percentage of restaurants offering online delivery
    delivery_counts = df['Has Online delivery'].value_counts()
    delivery_percentage = (delivery_counts / len(df)) * 100

    # Compute average ratings
    df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')  # Convert to numeric, handling errors
    avg_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()

    # Print results
    print("\nPercentage of restaurants offering online delivery:")
    print(delivery_percentage)
    print("\nAverage ratings:")
    print(avg_ratings)

    # Plot the number of restaurants offering online delivery
    plt.figure(figsize=(8, 6))
    ax = sns.barplot(x=delivery_counts.index, y=delivery_counts.values, palette='coolwarm')
    plt.title("Number of Restaurants Offering Online Delivery")
    plt.xlabel("Online Delivery")
    plt.ylabel("Count")
    
    # Annotate bars with actual counts
    for p, count in zip(ax.patches, delivery_counts):
        ax.annotate(f'{count}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=12, fontweight='bold')

    plt.show()

    # Print average ratings for both categories
    print("\nAverage Ratings:")
    for category, rating in avg_ratings.items():
        print(f"{category}: {rating:.2f}")
```
